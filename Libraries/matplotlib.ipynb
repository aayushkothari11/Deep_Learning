```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('PLOT')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.Figure()
```

```python
fig
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig, axes = plt.subplots(figsize=(10,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, 'r')
```

```python
fig.savefig("filename.png")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color="red", linewidth=3,alpha=0.5)
```

```python
 fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color="red", linewidth=3,alpha=0.5,marker='o', markersize=8, markerfacecolor="red")
```
